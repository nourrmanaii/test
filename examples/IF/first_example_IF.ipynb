```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
n_samples = 1000
n_features = 2
X = np.random.randn(n_samples, n_features)
```

```python

plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Apply Isolation Forest
clf = IsolationForest(contamination=0.1)  # Contamination defines the expected proportion of outliers
clf.fit(X)

# Predict outliers
y_pred = clf.predict(X)
```

```python
# Visualize the results
plt.scatter(X[:, 0], X[:, 1], c='blue', label='Normal')
plt.scatter(X[y_pred == -1, 0], X[y_pred == -1, 1], c='red', label='Outlier')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Isolation Forest Results')
plt.legend()
plt.show()
```
